```python
import numpy as np

a = np.arange(1, 11)
print(a)
```

```python
a = a.reshape((2, 5))
print(a)
```

```python
import matplotlib.pyplot as plt

plt.plot(a)
plt.show()
```

```python
# 1부터 100까지 7로 나눈 나머지가 4인 애들만 다 모아서 리스트로 만들어봐
# 코드길이는 한줄로 끝나야돼

a = [i for i in range(1, 101) if i % 7 == 4]
print(a)
```
