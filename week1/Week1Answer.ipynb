```python
import numpy as np

# 1000개의 랜덤 데이터 생성하기
# high 값은 범위에 안들어갑니다
data = np.random.randint(low=155, high=196, size=1000)
print('평균 = {}, 표준편차 = {}'.format(np.average(data), np.std(data)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(155, 196, 1000)

# 히스토그램을 그리기 위해서는 plt.hist 함수를 이용합니다!
plt.figure()
plt.hist(data)
plt.xlabel('Height of people')
plt.ylabel('Distribution')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 평균과 표준편차는 1번의 데이터 값을 가져와서
# 정규분포를 따르는 10000개의 데이터 셋을 생성해보겠습니다.
mu = 174.998
sigma = 11.501478
data = np.random.normal(mu, sigma, 10000)

plt.figure()
plt.hist(data, 50, histtype='stepfilled', normed=1)
plt.xlabel('Height of people')
plt.ylabel('Probability')
plt.grid()
plt.show()
```
